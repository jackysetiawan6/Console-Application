```python
from Bio.Seq import Seq as seq
from Bio.SeqUtils import molecular_weight as mw
from Bio.SeqUtils import MeltingTemp as mt
from matplotlib import pyplot as plt

sequence_A = seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
sequence_B = seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')

print(f'Molecular Weight A: {mw(sequence_A)}')
print(f'Wallace Melting Temperature A: {mt.Tm_Wallace(sequence_A)}')
print()
print(f'Molecular Weight B: {mw(sequence_B)}')
print(f'Wallace Melting Temperature B: {mt.Tm_Wallace(sequence_B)}')

plt.title('Melting Temperature')
plt.bar(['A', 'B'], [mt.Tm_Wallace(sequence_A), mt.Tm_Wallace(sequence_B)], color=['red', 'orange'])
plt.xlabel('Sequence')
plt.ylabel('Wallace Melting Temperature')
plt.show()

GC_A = (sequence_A.count('G') + sequence_A.count('C')) / len(sequence_A) * 100
GC_B = (sequence_B.count('G') + sequence_B.count('C')) / len(sequence_B) * 100
AT_A = (sequence_A.count('A') + sequence_A.count('T')) / len(sequence_A) * 100
AT_B = (sequence_B.count('A') + sequence_B.count('T')) / len(sequence_B) * 100
print(f'GC Percentage A: {GC_A} and AT Percentage A: {AT_A}')
print(f'GC Percentage B: {GC_B} and AT Percentage B: {AT_B}')

plt.title('GC and AT Content Percentage')
plt.bar(['GC_A', 'AT_A', 'GC_B', 'AT_B'], [GC_A, AT_A, GC_B, AT_B], color=['red', 'red', 'orange', 'orange'])
plt.xlabel('Sequence')
plt.ylabel('Percentage (%)')
for index, value in enumerate([GC_A, AT_A, GC_B, AT_B]):
    plt.text(index, value + 0.5, '{:.2f}'.format(value), ha='center')
plt.show()
```

```python
from Bio.Seq import Seq as seq

sequence_A = seq('ATGCGTTTATGACGTATCTGAAGTCTACTAGGGTGATCTGACTAA')
mRNA_A = sequence_A.transcribe()
amino_A = mRNA_A.translate()
complement_A = sequence_A.complement()
print('Sequence A')
print('DNA       : ', sequence_A)
print('mRNA      : ', mRNA_A)
print('amino     : ', amino_A)
print('complement: ', complement_A)

print()

sequence_B = seq('GTGTTAGTGCGCTAGTGCTGACTAGGTACAGTAGCTAGTTACTAG')
mRNA_B = sequence_B.transcribe()
amino_B = mRNA_B.translate()
complement_B = sequence_B.complement()
print('Sequence B')
print('DNA       : ', sequence_B)
print('mRNA      : ', mRNA_B)
print('amino     : ', amino_B)
print('complement: ', complement_B)

print()

sequence_C = seq(mRNA_A[:24] + mRNA_B[-21:])
amino_C = sequence_C.translate()
complement_C = sequence_C.complement()
print('Sequence C: ', sequence_C)
print('amino     : ', amino_C)
print('complement: ', complement_C)

print()

sequence_C_DNA = sequence_C.back_transcribe()
print('Sequence C (DNA): ', sequence_C_DNA)
```
